## Missing data :( 

```python
data = pd.read_csv('../data/mn-budget-detail-2014_nan.csv')
```

```python
len(data)
```

```python
data[data["amount"].isna()]
```

```python
data_dropped = data.dropna()
```

```python
data_dropped[data_dropped['amount'].isna()]
```

```python
len(data_dropped)
```

```python
# fill the missing amount by column mean
data['amount'].fillna(data['amount'].mean())
```

```python
# more accurate would be replace with category wise mean
data.groupby('category').transform(lambda x: x.fillna(x.mean()))
```

## Read data from different sources

```python
# This is just awesome
tables = pd.read_html('http://www.fdic.gov/bank/individual/failed/banklist.html')
df = tables[0]
```

```python
df.shape
```

```python
df.head()
```

```python
df.pivot_table(index=["ST", "City"], aggfunc='count')
```

```python
df.groupby(by=["ST", "City"]).count()
```

```python
# convert the data type
df["Closing Date"] = pd.to_datetime(df["Closing Date"])
```

```python
df.head()
```

```python
# "Closing date" is of type datetime now
df.info()
```

```python
df["Year"] = df["Closing Date"].apply(lambda x: x.year)
```

```python
df.head()
```

```python
df.groupby(by='Year')["Bank Name"].count()
```

```python
# We may not always need the next column to be created
df.groupby(df["Closing Date"].dt.year)["Bank Name"].count()
```
