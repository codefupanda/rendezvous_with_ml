# Viz

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Matplot lib

```python
%matplotlib inline
```

```python
plt.rcParams['figure.dpi'] = 100
```

```python
x = np.linspace(1, 10, num=20)
x
```

```python
y = x ** 2
y
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes.plot(x, y, 'b.-')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')


axes2.plot(x, x ** 3, 'g--')
```

```python
# multiple plots
fig, axes = plt.subplots(1, 2)

axes[0].plot(x, x**2, 'b.-')
axes[0].set_xlabel("X")
axes[0].set_ylabel("X ** 2")


axes[1].plot(x, x**3, 'g--')
axes[1].set_xlabel("X")
axes[1].set_ylabel("X ** 3")

plt.tight_layout()
```

```python
fig, axes = plt.subplots()

axes.scatter(np.random.rand(10), np.random.rand(10), c='r', marker='*')
```

```python
import seaborn as sns
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"], bins=40)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind = 'reg')
```

```python
sns.pairplot(tips, hue='sex')
```

### Categorical plots

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

### Matrix plots

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flts = flights.pivot_table(index='month', columns='year')
```

```python
flts.head()
```

```python
sns.heatmap(flts, cmap='YlGnBu')
```

```python
sns.clustermap(flts, cmap='YlGnBu')
```
